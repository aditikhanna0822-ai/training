```python
# q9
import csv
file = "b.csv"
while True:
    print("\n1 Add Book")
    print("2 Search Book")
    print("3 Delete Book")
    print("4 Exit")
    choice = input("Enter choice: ")
    if choice == "1":
        bookid = input("Enter book id: ")
        title = input("Enter title: ")
        author = input("Enter author: ")
        with open(file, "a", newline="") as f:
            writer = csv.writer(f)
            writer.writerow([bookid, title, author])
        print("Book added successfully")
    elif choice == "2":
        bookid = input("Enter book id to search: ")
        try:
            found = False
            with open(file, "r") as f:
                reader = csv.reader(f)
                for row in reader:
                    if row[0] == bookid:
                        print("Book id:", row[0])
                        print("Title:", row[1])
                        print("Author:", row[2])
                        found = True
                        break
            if not found:
                print("Book not found")
        except FileNotFoundError:
            print("File does not exist")
    elif choice == "3":
        bookid = input("Enter book id to delete: ")
        try:
            rows = []
            found = False
            with open(file, "r") as f:
                reader = csv.reader(f)
                for row in reader:
                    if row[0] == bookid:
                        found = True
                    else:
                        rows.append(row)
            if found:
                with open(file, "w", newline="") as f:
                    writer = csv.writer(f)
                    writer.writerows(rows)
                print("Book deleted successfully")
            else:
                print("Book not found")
        except FileNotFoundError:
            print("File does not exist")
    elif choice == "4":
        print("Exiting...")
        break
    else:
        print("Invalid choice")
```

```python
#q10
file = "expense.txt"
# Add expense
def add_expense():
    try:
        category = input("Enter category: ")
        amount = float(input("Enter amount: "))
        with open(file, "a") as f:
            f.write(category + "," + str(amount) + "\n")
        print("Expense added successfully")
    except ValueError:
        print("Invalid amount")
# View summary
def summary():
    try:
        totals = {}
        with open(file, "r") as f:
            for line in f:
                category, amount = line.strip().split(",")
                amount = float(amount)
                if category in totals:
                    totals[category] += amount
                else:
                    totals[category] = amount
        print("\nExpense Summary")
        for category in totals:
            print(category, ":", totals[category])
    except FileNotFoundError:
        print("File does not exist")
# Main program
while True:
    print("\n1 Add Expense")
    print("2 Show Summary")
    print("3 Exit")
    choice = input("Enter choice: ")
    if choice == "1":
        add_expense()
    elif choice == "2":
        summary()
    elif choice == "3":
        print("Exiting...")
        break
    else:
        print("Invalid choice")
```

```python
# q 11
import random
import string
class PasswordError(Exception):
    pass
# Generate random password
chars = string.ascii_letters + string.digits + string.punctuation
print("Generated Password:", "".join(random.choice(chars) for i in range(8)))
password = input("Enter password: ")
try:
    if len(password) < 8:
        raise PasswordError("Minimum 8 characters")
    if not any(i.isupper() for i in password):
        raise PasswordError("Add uppercase letter")
    if not any(i.islower() for i in password):
        raise PasswordError("Add lowercase letter")
    if not any(i.isdigit() for i in password):
        raise PasswordError("Add a number")
    if not any(i in string.punctuation for i in password):
        raise PasswordError("Add special character")
    print("Valid Password")
except PasswordError as e:
    print(e)
```

```python
#q 15
# Exam Result Processing System
students = []   # list for student records
def calculate_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 75:
        return "B"
    elif avg >= 60:
        return "C"
    elif avg >= 40:
        return "D"
    else:
        return "F"
n = int(input("Enter number of students: "))
for i in range(n):
    roll = int(input("Enter Roll No: "))
    name = input("Enter Name: ")
    student = (roll, name)      # tuple
    marks = {}                  # dictionary
    marks["Python"] = int(input("Python Marks: "))
    marks["DBMS"] = int(input("DBMS Marks: "))
    marks["Java"] = int(input("Java Marks: "))
    total = sum(marks.values())
    avg = total / 3
    grade = calculate_grade(avg)
    students.append([student, total, grade])
# Rank List
students.sort(key=lambda x: x[1], reverse=True)
print("\n----- RANK LIST -----")
rank = 1
for s in students:
    print("Rank:", rank)
    print("Roll No:", s[0][0])
    print("Name:", s[0][1])
    print("Total:", s[1])
    print("Grade:", s[2])
    print()
    rank += 1
# Save Result in File
file = open("result.txt", "w")
rank = 1
for s in students:
    file.write("Rank: " + str(rank) + "\n")
    file.write("Roll No: " + str(s[0][0]) + "\n")
    file.write("Name: " + s[0][1] + "\n")
    file.write("Total Marks: " + str(s[1]) + "\n")
    file.write("Grade: " + s[2] + "\n\n")
    rank += 1
file.close()
print("Results saved in result.txt")
```

```python
# q 17
import requests
from bs4 import BeautifulSoup
import csv

# Website URL
url = "https://quotes.toscrape.com/"

# Send request
response = requests.get(url)

# Check if website opened successfully
if response.status_code == 200:

    # Parse HTML
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all quotes
    quotes = soup.find_all("div", class_="quote")

    # Create CSV file
    with open("quotes.csv", "w", newline="", encoding="utf-8") as file:

        writer = csv.writer(file)

        # Header
        writer.writerow(["Quote", "Author"])

        # Write data
        for quote in quotes:
            text = quote.find("span", class_="text").text
            author = quote.find("small", class_="author").text

            writer.writerow([text, author])

    print("Data saved successfully in quotes.csv")

else:
    print("Failed to connect to website.")
```

```python
# q 20
import pandas as pd

# Load student data
df = pd.read_csv("students.csv")

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Print column names
print("Columns in CSV:", df.columns.tolist())

# Change these names according to your CSV
# Example: Maths -> Math
df.rename(columns={
    "Maths": "Math",
    "math": "Math",
    "science": "Science",
    "english": "English"
}, inplace=True)

# Check required columns
required = ["Math", "Science", "English"]

if not all(col in df.columns for col in required):
    print("\nError: Required columns not found.")
    print("Available columns are:", df.columns.tolist())
else:
    # Calculate Total and Percentage
    df["Total"] = df[required].sum(axis=1)
    df["Percentage"] = (df["Total"] / 300) * 100

    # Assign Grades
    def get_grade(p):
        if p >= 90:
            return "A+"
        elif p >= 80:
            return "A"
        elif p >= 70:
            return "B"
        elif p >= 60:
            return "C"
        else:
            return "F"

    df["Grade"] = df["Percentage"].apply(get_grade)

    # Display report
    print("\n===== STUDENT MARKS REPORT =====")
    print(df)

    # Statistics
    print("\nAverage Percentage:", round(df["Percentage"].mean(), 2))
    print("Highest Percentage:", round(df["Percentage"].max(), 2))
    print("Lowest Percentage:", round(df["Percentage"].min(), 2))

    # Topper
    topper = df.loc[df["Percentage"].idxmax()]
    print("\nTopper:")
    print(topper)

    # Save report
    df.to_csv("student_report.csv", index=False)
    print("\nReport saved as student_report.csv")
```
